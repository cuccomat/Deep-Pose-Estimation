```python
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

#matplotlib.rcParams.update({'font.size': 18, 'text.usetex': True})
```

```python
frist_rot = pd.read_csv('drive/Othercomputers/Il mio laptop/Deep-Pose-Estimation/baseline/plots/UNET Loss.csv')

frist_rot.head()
```

```python
# Rotation loss
frist_rot = frist_rot.plot(x='Step', y= ['good-cloud-92 - rot_loss [°]', 'good-cloud-92 - rot_val_loss [°]'],figsize=(6,4.3), linewidth=0.5, fontsize = 14.5, logy = True)

frist_rot.set_ylabel("$Loss\ [^{\circ}]$", fontsize=16)
frist_rot.set_xlabel("$Step$", fontsize=16)
frist_rot.legend(["$rot\ loss$", "$rot\ val\ loss$"], fontsize=14.5)

frist_rot.figure.savefig('9th_rot.pgf')
```

```python
# Translations loss
frist_rot = frist_rot.plot(x='Step', y= ['good-cloud-92 - tra_loss [cm]', 'good-cloud-92 - tra_val_loss [cm]'],figsize=(6,4.3), linewidth=0.5, fontsize = 14.5, logy = True)

frist_rot.set_ylabel("$Loss\ [cm]$", fontsize=16)
frist_rot.set_xlabel("$Step$", fontsize=16)
frist_rot.legend(["$tra\ loss$", "$tra\ val\ loss$"], fontsize=14.5)

frist_rot.figure.savefig('9th_tra.pgf')
```

```python
# Global loss
frist_rot = frist_rot.plot(x='Step', y= ['bright-waterfall-38 - loss', 'bright-waterfall-38 - val_loss'],figsize=(6,4.3), linewidth=0.5, fontsize = 14.5, logy = True)

frist_rot.set_ylabel("$Loss$", fontsize=16)
frist_rot.set_xlabel("$Step$", fontsize=16)
frist_rot.legend(["$global\ loss$", "$global\ val\ loss$"], fontsize=14.5)




frist_rot.figure.savefig('UNET_LOSS.pgf')
```
